```python
import pandas as pd
import numpy as np
```

```python
student_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
student_performance = student_performance.rename(columns = 
                          {'math score' : 'math_score',
                          'reading score' : 'reading_score',
                          'writing score' : 'writing_score'})
```

```python
student_performance.head()
```

```python
student_performance.groupby('gender').mean()
```

```python
#в aggregate передаются переменные, которые мы хотим использовать для аггрегации
#и что мы хотим для этих переменных узнать
student_performance.groupby('gender', as_index = False).aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
                                    .rename(columns = 
                                           {'math_score' : 'math_score_mean',
                                           'reading_score' : 'reading_score_mean'})
```

```python
#Группировка по нескольким столбцам, как вывод получим все пересечения этих столбцов
student_performance.groupby(['gender','race/ethnicity'], as_index = False).aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
                                    .rename(columns = 
                                           {'math_score' : 'math_score_mean',
                                           'reading_score' : 'reading_score_mean'})
```

```python
#Группировка по нескольким столбцам, как вывод получим все пересечения этих столбцов
mean_scores = student_performance.groupby(['gender','race/ethnicity']).aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
                                    .rename(columns = 
                                           {'math_score' : 'math_score_mean',
                                           'reading_score' : 'reading_score_mean'})
```

```python
#Сохраним сумму баллов по всем предметам в новую пандаскую серию total_score
student_performance['total_score'] = student_performance.math_score + student_performance.reading_score + student_performance.writing_score
```

```python
#При сложении пандаских серий получается новая пандаская серия, которая содержит в себе сумму всех указанных пандаских серий
student_performance.math_score + student_performance.reading_score
```

```python
student_performance
```

```python
#Данный способ позволяет за раз создать только одну колонку 
student_performance['total_score'] = student_performance.math_score + student_performance.reading_score + student_performance.writing_score
```

```python
#Применяя логарифм к серии мы применяем логарифм к каждому отдельному элементу
np.log(student_performance.total_score)
```

```python
#Данный спосо позволяет за раз создать несколько колонок
student_performance = student_performance.assign(total_score_log = np.log(student_performance.total_score))
```

```python
#Для удаления колонок используется функция drop
#axis = 1 указывает на то что работаем с колонкой, а не со строкой
student_performance.drop(['total_score'], axis = 1)
```
