```python
# axis = 0 фильтрует строки, axis = 1 фильтрует столбцы
student_performance_with_names.filter(like = 'e', axis = 0) 
```

```python
#items отбирает колонки по названию
student_performance_with_names.filter(items = ['Cersei'], axis = 0)
```

```python
# regex выводит колонки фильтруя по регулярному выражению
# ^- совпадение в начале строки, $ - совпадение в конце строки
# ^g - выведет колонки, начинающиеся с g
# e$ - выведет колонки, заканичвающиеся на e
student_performance_with_names.filter(regex = 'e$', axis = 1)
```

```python
# О filter
#1.Можно использовать толко один из аргументов items, like, regex
#2. Аргумент regex отбирает лэйблы, подходящие к переданному паттерну
#3. Аргумент like отбирает те лэйблы, где встречается поданное значение
#4. Аргумент axis отвечает за то, какие лэйблы мы рассматриваем - индекс или колонки
#5. Аргумент items отбирает лэйблы, переданные в коллекции в качестве значения
```
