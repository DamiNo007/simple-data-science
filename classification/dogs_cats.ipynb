```python
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import cv2
from google.colab import files
```

```python
train, _ = tfds.load('cats_vs_dogs', split = ['train[:100%]'], with_info = True, as_supervised = True)
```

```python
SIZE = 224

def img_resize(img, label):
    img = tf.cast(img, tf.float32)
    img = tf.image.resize(img, (SIZE, SIZE))
    img = img / 255.0
    return img, label
```

```python
train_resized = train[0].map(img_resize)
train_batches = train_resized.shuffle(1000).batch(16)
```

```python
train
```

```python
base_layers = tf.keras.applications.MobileNetV2(input_shape = (SIZE, SIZE, 3), include_top = False)
base_layers.trainable = False
```

```python
#Dropout() позволяет внезапно выключать некоторые нейроны, чтобы нейросеть не запоминала данные
model = tf.keras.Sequential([
    base_layers, 
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile(optimizer = 'adam', loss = tf.keras.losses.BinaryCrossentropy(from_logits = True), metrics = ['accuracy'])
```

```python
model.fit(train_batches, epochs = 3)
```

```python
img = cv2.imread('C:/Users/Asus/Desktop/NEURAL_NETWORKS/Convolutional NN/cats vs dogs/img/bulldog.jpg')
img_arr = np.array(img)
img_resized, _ = img_resize(img_arr, _)
img_expanded = np.expand_dims(img_resized, axis = 0)
prediction = model.predict(img_expanded)
if prediction < 0.5:
    print("CAT: %f percent" % (100 - prediction * 100))
else: print ("DOG: %f percent" % (prediction * 100))
```
